## 1. RAG 파이프라인 개요
- Load Data(데이터를 읽어옴) - Text Split(작은 청크 단위로 분할) - Indexing(분할된 각각의 청크 단위들을 벡터 스토어로 인덱싱) - Retrieval(검색 단계) - Generation(생성 단계)



##### Step 1: Load Data

```python
# Data Loader - 웹페이지 데이터 가져오기
from langchain_community.document_loaders import WebBaseLoader

# 위키피디아 정책과 지침
url = 'https://ko.wikipedia.org/wiki/%EC%9C%84%ED%82%A4%EB%B0%B1%EA%B3%BC:%EC%A0%95%EC%B1%85%EA%B3%BC_%EC%A7%80%EC%B9%A8'
loader = WebBaseLoader(url) # url에 해당되는 웹 페이지의 텍스트 데이터를 불러옴

# 웹페이지 텍스트 -> Documents
docs = loader.load() # 웹페이지의 문서들을 list형태로 docs에 저장

print(len(docs))
print(len(docs[0].page_content))
print(docs[0].page_content[5000:6000])
```

##### Step 2: Split texts

```python
# Text Split (Documents -> small chunks: Documents)
from langchain.text_splitter import RecursiveCharacterTextSplitter # 단락이나 문장, 구 단위로 텍스트를 작은 조각으로 나눠주는 클래스

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200) # 각 조각(chunk)의 크기는 1000글자, 각 조각들은 200자 정도 중복됨
splits = text_splitter.split_documents(docs) #docs라는 리스트에 있는 document 객체들을 작은 chunk 단위로 조각내어 splits에 저장

print(len(splits))
print(splits[10])
```

```python
# page_content 속성
splits[10].page_content # RecursiveCharacterTextSplitter를 이용했기 때문에 문장의 중간에서 split되는게 아니라, 줄바꿈 문자나 단락, 문장 또는 쉼표, 마침표 기준으로 나눠지기 때문에 문장의 끝부분이 온전한 형태로 마무리됨
```

```python
# metadata 속성
splits[10].metadata
```

##### Step 3: Indexing

```python
# 인덱싱을 하기 위해 필요한 2가지
# 1. 앞서 조각낸 chunk들을 벡터 임베딩으로 변환해주는 작업
# 2. 숫자 벡터 형태로 변환된 임베딩 벡터를 벡터 스토어에 저장

# Indexing (Texts -> Embedding -> Store)
from langchain_community.vectorstores import Chroma #벡터 스토어(Chroma DB)
from langchain_openai import OpenAIEmbeddings # 임베딩 모델

# from_documents : OpenAI 임베딩 모델을 활용하여 각각의 chunk(splits)들을 벡터로 변환하여 벡터 스토어(Chroma DB)에 저장
vectorstore = Chroma.from_documents(documents=splits, # 어떤 문서를 임베딩할 것인지
                                    embedding=OpenAIEmbeddings()) # 어떤 임베딩 모델을 사용할 것인지

# similarity_search : 질문과 가장 유사도가 높은 문서 조각(chunk)를 벡터 스토어에서 찾아온다.
docs = vectorstore.similarity_search("격하 과정에 대해서 설명해주세요.")
print(len(docs))
print(docs[0].page_content)
```

##### Step 4: Retrieval ~ Generation

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser

# Prompt
template = '''Answer the question based only on the following context:
{context}

Question: {question}
'''

prompt = ChatPromptTemplate.from_template(template)

# LLM
model = ChatOpenAI(model='gpt-3.5-turbo-0125', temperature=0)

# Retriever(검색 도구 - 여기서는 벡터 스토어로 선언)
retriever = vectorstore.as_retriever()

# Combine Documents
def format_docs(docs):
    return '\n\n'.join(doc.page_content for doc in docs)

# RAG Chain 연결
rag_chain = (
    {'context': retriever | format_docs, 'question': RunnablePassthrough()} # retriever를 통해 검색해온 여러 개의 문서들을 combine해서 context에 넣어줌
    | prompt
    | model # 답변 생성
    | StrOutputParser()
)

# Chain 실행
rag_chain.invoke("격하 과정에 대해서 설명해주세요.")
```
